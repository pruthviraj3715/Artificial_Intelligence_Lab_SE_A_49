```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# -------------------------------
# 1. Generate Normal Distribution Data
# -------------------------------

# Parameters
mu = 50        # Mean
sigma = 10     # Standard deviation
size = 1000    # Number of samples

# Generate random data
data = np.random.normal(mu, sigma, size)

# -------------------------------
# 2. Descriptive Statistics
# -------------------------------

mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)

print("Descriptive Statistics:")
print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)

# -------------------------------
# 3. Plot Normal Distribution
# -------------------------------

plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, sigma)

plt.plot(x, p, 'k', linewidth=2)
plt.title("Normal Distribution (Mean = 50, Std Dev = 10)")
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.show()

# -------------------------------
# 4. Coin Toss Simulation
# -------------------------------

tosses = 1000
coin = np.random.choice(['H', 'T'], size=tosses)

heads = np.sum(coin == 'H')
tails = np.sum(coin == 'T')

print("\nCoin Toss Simulation:")
print("Total Tosses:", tosses)
print("Heads:", heads)
print("Tails:", tails)
print("Probability of Heads:", heads/tosses)
print("Probability of Tails:", tails/tosses)
```
