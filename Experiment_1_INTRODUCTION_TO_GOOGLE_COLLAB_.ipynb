###Program to compute factorial of a given number

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = int(input("Enter a number: "))
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

###Program to display table for a given number

```python
def display_multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")

# Example usage:
number_for_table = int(input("Enter a number to see its multiplication table: "))
display_multiplication_table(number_for_table)
```

###Program to identify the given number is odd or even

```python
num = int(input("Enter a number: "))

if (num % 2) == 0:
   print(f"{num} is an even number")
else:
   print(f"{num} is an odd number")
```

### Program to identify if the entered number is palindrome or not.

```python
num = int(input("Enter a number: "))

# Convert number to string for easy reversal
num_str = str(num)

# Reverse the string
reversed_num_str = num_str[::-1]

# Check if the original and reversed strings are the same
if num_str == reversed_num_str:
    print(f"{num} is a palindrome number")
else:
    print(f"{num} is not a palindrome number")
```

### Decision Rules for Stream Recommendation Based on Favorite Subject




#### Implement Decision Logic (Python If-Else with Input)





```python
favorite_subject = input("Enter your favorite subject: ").lower()

if favorite_subject in ['physics', 'chemistry', 'biology', 'computer science']:
    print('Recommended Stream: Science')
elif favorite_subject in ['accountancy', 'business studies', 'economics']:
    print('Recommended Stream: Commerce')
elif favorite_subject == 'mathematics':
    print('Recommended Stream: Science (Mathematics is foundational for Science, but also important for Commerce)')
else:
    print('Recommended Stream: Arts/Humanities (Based on your favorite subject, this stream allows for diverse studies).')
```
